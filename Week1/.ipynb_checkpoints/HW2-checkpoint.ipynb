```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
rawdata = np.fromfile('lena_gray512.raw',dtype=np.uint8)
```

```python
rawdata.shape
```

```python
img = rawdata.reshape([512,512])
```

```python
img.shape
```

```python
plt.imshow(img,cmap='gray')
plt.show()
```

```python
h, w = 512, 512
```

```python
nh, nw = 1024, 1024
```

```python
scale = int(nh/h)
```

```python
nimg = np.zeros([1024,1024])
```

```python
for ni in range(0,nh,2):
    for nj in range(0,nw,2):
        i = int(ni/scale)
        j = int(nj/scale)
        nimg[ni,nj] = img[i,j]
        nimg[ni,nj+1] = img[i,j]
        nimg[ni+1,nj] = img[i,j]
        nimg[ni+1,nj+1] = img[i,j]
```

```python
plt.imshow(nimg,cmap='gray')
plt.show()
```
