```python
imgHeight=512
imgWidth=512
newimgHeight=1024
newimgWidth=1024
imgScale=int(newimgHeight/imgHeight)
newimage=np.zeros((1024,1024))
for i_new in range(0, newimgHeight,2):
    for j_new in range(0, newimgWidth, 2):
        i=int(i_new/imgScale)
        j=int(j_new/imgScale)
        #print("i:", i)
        #print("j:", j)
        #print("i_new:", i_new)
        #print("j_new:", j_new)
        newimage[i_new,j_new] = image[i,j]
        newimage[i_new,j_new+1] = image[i,j]
        newimage[i_new+1,j_new] = image[i,j]
        newimage[i_new+1,j_new+1] = image[i,j]
```
