```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# img = cv.imread('city.jpeg')
img = cv.imread('grayworld_assumption_1.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
print('R_min: ',img[:,:, 0].min(), '; R_max: ', img[:,:,0].max())
print('G_min: ',img[:,:, 1].min(), '; G_max: ', img[:,:,1].max()) 
print('B_min: ',img[:,:, 2].min(), '; B_max: ', img[:,:,2].max()) 
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
norm_img = np.zeros((800,800))
final_img = cv.normalize(img,  norm_img, 0, 255, cv.NORM_MINMAX)
plt.imshow(final_img)
```

```python
print('R_min: ',final_img[:,:, 0].min(), '; R_max: ', final_img[:,:,0].max())
print('G_min: ',final_img[:,:, 1].min(), '; G_max: ', final_img[:,:,1].max()) 
print('B_min: ',final_img[:,:, 2].min(), '; B_max: ', final_img[:,:,2].max()) 
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([final_img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
